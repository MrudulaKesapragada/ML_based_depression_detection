```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('b_depressed.csv')
df
```

```python
df.columns
```

```python
df.dropna(inplace = True)
df
```

```python
X = df.drop(['Survey_id', 'Ville_id', 'depressed'], axis = 1)
y = df['depressed']
```

# Train - Test - Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression(random_state = 0)
model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_test)

# Accuracy
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))

# Confusion Matrix
from sklearn.metrics import confusion_matrix
plt.figure(figsize = (10,10))
sns.heatmap(confusion_matrix(y_test, y_pred), cmap = "Blues", annot = True, fmt = '')
```

# Support Vector Classifier

```python
from sklearn.svm import SVC
model_svm = SVC(kernel='rbf')
model_svm.fit(X_train, y_train)
y_pred = model_svm.predict(X_test)

# Accuracy
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize = (10,10))
sns.heatmap(confusion_matrix(y_test, y_pred), cmap = "Blues", annot = True, fmt = '')
```

# Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
model_mnb = MultinomialNB()
model_mnb.fit(X_train, y_train)
y_pred = model_mnb.predict(X_test)

# Accuracy
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize = (10,10))
sns.heatmap(confusion_matrix(y_test, y_pred), cmap = "Blues", annot = True, fmt = '')
```

# Multi-Layer Perceptron (MLP) Classifier

```python
from sklearn.neural_network import MLPClassifier
model_mlp = MLPClassifier(activation = 'logistic', hidden_layer_sizes = [20, 10, 5, 2])
model_mlp.fit(X_train, y_train)
y_pred = model_mlp.predict(X_test)

# Accuracy
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize = (10,10))
sns.heatmap(confusion_matrix(y_test, y_pred), cmap = "Blues", annot = True, fmt = '')
```
