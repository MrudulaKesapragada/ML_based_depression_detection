```python
import pickle
import time
import warnings
warnings.filterwarnings('ignore')
from pywebio.input import *
from pywebio.output import *
from pywebio.platform.flask import webio_view
from flask import Flask
app = Flask(__name__)

def getStarted():
    # with open('depressed_data', 'rb') as f:
    #     df = pickle.load(f)
    with open('model_svm', 'rb') as f:
        model_svm = pickle.load(f)
    put_markdown('# Depression Detection')
    put_image(open('depression.jpg', 'rb').read(), width='100%')
    option = radio('Select a suitable option', options = ['Exploratory Data Analysis (EDA) of the depression data', 'Predict depressed or not'], required = True)
    if(option == 'Exploratory Data Analysis (EDA) of the depression data'):
        put_html('<hr>')
        put_markdown('### Please wait! Your request is being processed. ')
        put_processbar('bar');
        for i in range(1, 11):
            set_processbar('bar', i / 10)
            time.sleep(0.1)
        put_markdown('# Exploratory Data Analysis (EDA) of the depression data')
        
    if(option == 'Predict depressed or not'):
        sex = select('What is your sex?', options = ['Male', 'Female'], required = True)
        if(sex == 'Male'):
            sex = 1.0
        else:
            sex = 0.0
        Age = input('What is your age?', type = FLOAT, required = True)
        Married = select('Are you married or not?', options = ['Yes', 'No'], required = True)
        if(Married == 'Yes'):
            Married = 1.0
        else:
            Married = 0.0
        Number_children = input('How many children do you have?', type = FLOAT, required = True)
        education_level = input('Share your education level', type = FLOAT, required = True)
        total_members = input('How many members live in your family including yourself?', type = FLOAT, required = True)
        gained_asset = input('How much is your gained asset in rupees?', type = FLOAT, required = True)
        durable_asset = input('How much is your durable asset in rupees?', type = FLOAT, required = True)
        save_asset = input('How much is your saved asset in rupees?', type = FLOAT, required = True)
        living_expenses = input('Share your living expenses', type = FLOAT, required = True)
        other_expenses = input('Share your other expenses', type = FLOAT, required = True)
        incoming_salary = radio('Are you a salaried employee?', options = ['Yes', 'No'], required = True)
        if(incoming_salary == 'Yes'):
            incoming_salary = 1.0
        else:
            incoming_salary = 0.0
        incoming_own_farm = radio('Do you own a farm?', options = ['Yes', 'No'], required = True)
        if(incoming_own_farm == 'Yes'):
            incoming_own_farm = 1.0
        else:
            incoming_own_farm = 0.0
        incoming_business = radio('Do you own a business?', options = ['Yes', 'No'], required = True)
        if(incoming_business == 'Yes'):
            incoming_business = 1.0
        else:
            incoming_business = 0.0
        incoming_no_business = input('How many businesses do you own?', type = FLOAT, required = True)
        incoming_agricultural = input('What is your agricultural income?', type = FLOAT, required = True)
        farm_expenses = input('Share your farm expenses', type = FLOAT, required = True)
        labor_primary = radio('Do you have salaried labours?', options = ['Yes', 'No'], required = True)
        if(labor_primary == 'Yes'):
            labor_primary = 1.0
        else:
            labor_primary = 0.0
        lasting_investment = input('Share lasting investment', type = FLOAT, required = True)
        no_lasting_investment = input('Share non-lasting investment', type = FLOAT, required = True)
        put_html('<hr>')
        put_markdown('### Please wait! Your request is being processed. ')
        put_processbar('bar');
        for i in range(1, 11):
            set_processbar('bar', i / 10)
            time.sleep(0.1)
        depressedOrNot = model_svm.predict([[sex, Age, Married, Number_children, education_level, total_members, gained_asset, durable_asset, save_asset, living_expenses, other_expenses, incoming_salary, incoming_own_farm, incoming_business, incoming_no_business, incoming_agricultural, farm_expenses, labor_primary, lasting_investment, no_lasting_investment]]) 
        if(depressedOrNot == 0):
            put_markdown('## Predicted Depression Status = `Depressed`')
        else:
            put_markdown('## Predicted Depression Status = `Not Depressed`')
    put_html('<hr>')
    put_html('<h1 style="font-size:20px; text-align:center; color:blue; background-color:powderblue;"> Thank you! :-) <br> </h1>') 
    put_html('<hr>')
app.add_url_rule('/depression', 'webio_view', webio_view(getStarted), 
                 methods=['GET', 'POST', 'OPTIONS'])
app.run(host='localhost', port=80)
# http://localhost/depression
```
